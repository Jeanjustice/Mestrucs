```python
# Importation des bibliothques ncessaires
import numpy as np
import matplotlib.pyplot as plt

# Gnration de donnes alatoires
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.randn(100) * 0.1

# Cration du graphique
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Donnes alatoires', color='green')
plt.title('Graphique des donnes alatoires')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Affichage du graphique
plt.show()
```
